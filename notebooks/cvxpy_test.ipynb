```python
import cvxpy as cp
import numpy
```

```python
# Problem data.
m = 2000
n = 1000
numpy.random.seed(1)
A = numpy.random.randn(m, n)
b = cp.Parameter(m)

# Construct the problem.
x = cp.Variable(n)
prob = cp.Problem(cp.Minimize(cp.sum_squares(A @ x - b)),
                   [x >= 0])
```

```python
b.value = numpy.random.randn(m)
#prob.solve()
prob.solve(solver=cp.MOSEK)
print(prob.value)
print("First solve time:", prob.solver_stats.solve_time)

b.value = numpy.random.randn(m)
#prob.solve(warm_start=True, verbose=True)
prob.solve(solver=cp.MOSEK, warm_start=True, verbose=True)
print(prob.value)
print("Second solve time:", prob.solver_stats.solve_time)
```
