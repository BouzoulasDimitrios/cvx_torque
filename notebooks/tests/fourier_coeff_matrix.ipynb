```python
def fft(time, torques):
    Fs = 1/np.mean(np.diff(time)) # Sampling frequency
    torques -= np.mean(torques)
    L = len(torques)
    Y = np.fft.fft(torques)
    P2 = abs(Y/L)
    P1 = P2[0:int(L/2)]
    P1[1:-2] = 2*P1[1:-2]
    f = (Fs/L)*np.linspace(0, (L/2), int(L/2))

    return f, P1
```

```python
t = np.arange(0, 10, 0.01)
u = np.sin(4*2*np.pi*t)
```

```python
plt.plot(t, u)
plt.show()
```

```python
freq, P = fft(t, u)
```

```python
plt.plot(freq, P)
plt.xlim(0, 10)
plt.show()
```
