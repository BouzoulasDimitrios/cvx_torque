## Robust two-stage Kalman filters for systems with unknown inputs

Chien-Shu Hsieh, "Robust two-stage Kalman filters for systems with unknown inputs," in IEEE Transactions on Automatic Control, vol. 45, no. 12, pp. 2374-2378, Dec. 2000, doi: 10.1109/9.895577. keywords: {Robustness;Filtering;State estimation;Kalman filters;Adaptive filters;Time varying systems;Covariance matrix;Lagrangian functions;Noise measurement;Stochastic processes},
Chien-Shu Hsieh, "Robust two-stage Kalman filters for systems with unknown inputs," in , vol. 45, no. 12, pp. 2374-2378, Dec. 2000, doi: 10.1109/9.895577. keywords: {Robustness;Filtering;State estimation;Kalman filters;Adaptive filters;Time varying systems;Covariance matrix;Lagrangian functions;Noise measurement;Stochastic processes}, 

```python
def robust_2stage_KF(m0, P0, y, N, A, E, H, Q, R):
    x = np.zeros((N,3))
    P = np.zeros((N,3,3))
    d = np.zeros(N)
    I = np.eye(3)
    
    x[0], P[0] = m0, P0
    
    for i in range(1, N):
        # State estimate prediction
        x_ = A @ x[i-1]
        P_ = A @ P[i-1] @ A.T + Q
        C_inv = la.inv(H @ P_ @ H.T + R)
        v = y[i] - H @ x_
        
        # Input estimate update
        P_d = la.inv(E.T @ H.T @ C_inv @ H @ E)
        K_d = P_d @ E.T @ H.T @ C_inv
        d[i] = K_d @ v
        
        # State estimate update
        K = P_ @ H.T @ C_inv
        V = (I - K @ H) @ E
        P_k = (I - K @ H) @ P_
        x_k = x_ + K @ v
        
        x[i] = x_k + (V * d[i]).ravel()
        P[i] = P_k + V @ P_d @ V.T
    
    return d, x, P
```

```python
# A
A_k = np.array([
    [0.1, 0.5, 0.08],
    [0.6, 0.01, 0.04],
    [0.1, 0.7, 0.05]
])

# C
H_k = np.array([
    [1, 1, 0],
    [0, 1, 1]
])

# B
E_k = np.array([
    [0],
    [2],
    [1]
])

Q_k = np.diag([10, 10, 10])
R_k = np.diag([20, 20])
```

```python
N = 40  # No. timesteps
ds = (50*(np.hstack((np.ones(N-20), -np.ones(N-20)))))  # input
```

```python
xs = np.zeros((N,3))
ys = np.zeros((N,2))

psd = la.cholesky(Q_k)
msd = la.cholesky(R_k)

for i in range(1,N):
    xs[i] = A_k @ xs[i-1] + (E_k * ds[i-1]).ravel() + psd @ np.random.normal(0, 1, *xs[i].shape)
    ys[i] = H_k @ xs[i] + msd @ np.random.normal(0, 1, *ys[i].shape)
```

```python
m0 = np.ones(3)
P0 = np.diag([10, 10, 10])
```

```python
inputs, states, covs = robust_2stage_KF(m0, P0, ys, N, A_k, E_k, H_k, Q_k, R_k)
```

```python
plt.plot(np.linspace(0, 40, 40), ds, label='True input')
plt.plot(np.linspace(0, 40, 40), inputs, '-', label='Estimated input')
plt.legend()
plt.xlabel("Timestep")
plt.ylabel("Input value")
plt.show()
```

```python
plt.subplot(221)
plt.plot(xs[:,0], label="State 1")
plt.plot(states[:,0], '-')
plt.legend()
plt.subplot(222)
plt.plot(xs[:,1], label="State 2")
plt.plot(states[:,1], '-')
plt.legend()
plt.subplot(223)
plt.plot(xs[:,2], label="State 3")
plt.plot(states[:,2], '-')
plt.legend()
plt.subplot(224)
plt.plot(ys, label="Observation")
plt.legend()
plt.show()
```

## Unbiased minimum-variance input and state estimation for linear discrete-time systems

https://doi.org/10.1016/j.automatica.2006.08.002

```python
def robust_recursive_KF(m0, P0, y, N, A, G, C, Q, R):
    x = np.zeros((N,3))
    P = np.zeros((N,3,3))
    d = np.zeros(N)
    I = np.eye(3)
    F = C @ G
    
    x[0], P[0] = m0, P0
    
    for i in range(1, N):
        # State estimate prediction
        x_ = A @ x[i-1]
        P_ = A @ P[i-1] @ A.T + Q
        
        R_k_inv = la.inv(C @ P_ @ C.T + R)
        
        # Input estimate update
        M_k = la.inv(F.T @ R_k_inv @ F) @ F.T @ R_k_inv
        d[i-1] = M_k @ (y[i] - C @ x_)
        
        # State estimate update
        K_k = P_ @ C.T @ R_k_inv
        x_k = x_ + (G * d[i-1]).ravel()
        x[i] = x_k + K_k @ (y[i] - C @ x_k)
        # Estimate covariance update
        P_k = (I - G @ M_k @ C) @ P_ @ (I - G @ M_k @ C).T + G @ M_k @ R @ M_k.T @ G.T
        S_k = -G @ M_k @ R
        P[i] = P_k - K_k @ (P_k @ C.T + S_k).T
    
    d[-1] = d[-2]
        
    return d, x, P
```

```python
inputs2, states2, covs2 = robust_recursive_KF(m0, P0, ys, N, A_k, E_k, H_k, Q_k, R_k)
```

```python
plt.plot(np.linspace(0, 40, 40), ds, label='True input')
plt.plot(np.linspace(0, 40, 40), inputs2, '-', label='Estimated input')
plt.legend()
plt.xlabel("Timestep")
plt.ylabel("Input value")
plt.show()
```

```python
plt.subplot(221)
plt.plot(xs[:,0], label="State 1")
plt.plot(states2[:,0], '-')
plt.legend()
plt.subplot(222)
plt.plot(xs[:,1], label="State 2")
plt.plot(states2[:,1], '-')
plt.legend()
plt.subplot(223)
plt.plot(xs[:,2], label="State 3")
plt.plot(states2[:,2], '-')
plt.legend()
plt.subplot(224)
plt.plot(ys, label="Observation")
plt.legend()
plt.show()
```
